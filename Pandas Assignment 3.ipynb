```python
import pandas as pd
```

```python
# import 1st Data Frame
df1 = pd.read_excel("File1.xlsx")
df1
```

```python
# import 2nd Data Frame
df2 = pd.read_excel("File2.xlsx")
df2
```

# Task 1: Merge two DataFrames with different columns

```python
pd.merge(df1, df2, how = "right", left_on = "CNIC", right_on = "NIC")
```

# Task 2: Merge two DataFrames on multiple columns

```python
pd.merge(df1, df2, how = "right", left_on = ["CNIC", "Account_no"], right_on = ["NIC", "Account_no"])
```

# Task 3: Merge two DataFrames on index

```python
# Following code performs an inner join by default, so only the indexes that appear in both Data Frames are kept 
pd.merge(df1, df2, left_index = True, right_index=True)
```
