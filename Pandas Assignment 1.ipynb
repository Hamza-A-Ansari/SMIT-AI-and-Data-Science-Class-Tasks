```python
indDate = pd.date_range("2022-08-20", periods=10)
# alphabet = list(string.ascii_uppercase)
alphabet = [chr(65+i) for i in range(26)]
```

```python
df2 = pd.DataFrame(np.random.randint(1, 2600, size = (len(indDate),len(alphabet))),
                   index = indDate,
                   columns = alphabet)
```

```python
df2.sort_values(by = ['X','Y'])
```

```python
# alphabet = [chr(65+i) for i in range(26)]
colArray = np.array(alphabet)
```

```python
tempList = []
for x in alphabet:
    tempList = []
    for y in alphabet:
        tempList.append(x+y)
    colArray = np.concatenate((colArray, tempList))
```

```python
df3 = pd.DataFrame(np.random.randint(1, 2600, size = (len(indDate),len(colArray))),
                   index = indDate,
                   columns = colArray)
df3.head()
```
